# Scraping Wikipedia - Step 1

Let's scrape our Wikipedia page using BeautifulSoup

```python
#wikipedia link
wiki = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

#open the page
page = urlopen(wiki)

#soup it up
soup = BeautifulSoup(page)
```

# Wikipedia Scraping - Part 2

Let's create some open lists that we can append Postcodes, Boroughs and Neighborhoods to as we scrape our Wikipedia page.  Then, let's take a look at what our table data looks like in Wikipedia.  

```python
postcode = []
borough = []
neighborhood = []

table = soup.find("table", { "class" : "wikitable sortable" })
print(table)
```

# Wikipedia Scraping - Part 3

Now let's loop over our Wikipedia page and append to our empty lists.  Then let's make sure that we remove the '\n' from our Neighborhoods.  Let's also check the lenght of each list to ensure we have an equal number of each.

```python
#scrape the zip codes, boroughs and neighborhoods

for row in table.findAll('tr'):
    cells = row.findAll('td')
    if len(cells)==3:
        postcode.append(cells[0].find(text=True))
        borough.append(cells[1].find(text=True))
        neighborhood.append(cells[2].find(text=True))

neighborhood_2 = []
for i in neighborhood:
    i = i.replace('\n', '')
    neighborhood_2.append(i)

print(postcode)
print(borough)
print(neighborhood_2)

print(len(postcode))
print(len(borough))
print(len(neighborhood_2))
```

# Creating DataFrame - Part 1

Let's create a DataFrame using the lists we created from our scrape

```python
#create a dataframe from lists

main_df = pd.DataFrame({"Postcode": postcode, "Borough": borough, "Neighborhood": neighborhood_2})
main_df.head()
```

# Creating DataFrame - Part 2

Let's create a copy just in case...

```python
main_df2 = main_df.copy()
main_df2
```

# Creating DataFrame - Part 3

Let's combine our rows where we have duplicate postal codes.

```python
main_df3 = main_df2['Neighborhood'].groupby(main_df2['Postcode']).apply(', '.join).reset_index()
main_df3
```

# Creating a DataFrame - Part 4

Let's create a DataFrame we can merge with our new de-duped DataFrame.

```python
main_posts = main_df[['Postcode', 'Borough']]
main_posts
```

# Creating DataFrame - Part 5

Let's remove any duplication that occurred during the merge process.

```python
new_main = pd.merge(main_df3, main_posts)
new_main2 = new_main.drop_duplicates()
new_main2
```

# Creating DataFrame - Part 6

Let's fix the value for Queen's Park Neighborhood to Queen's Park.

```python
new_main2.set_value(195, 'Neighborhood', "Queen's Park")
new_main2
```

# Creating DataFrame - Part 7

Let's drop all rows that have Not Assigned.

```python
new_main3 = new_main2[new_main2.Borough != 'Not assigned']
new_main3
```
